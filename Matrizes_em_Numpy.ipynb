**Estatística - Média**

```python
arr6 = np.array([23, 56, 78, 90])
arr6.mean()
```

**Desvio Padrão**

indica o quanto os dados estão afastados da média. é a raiz quadrada da variancia.

desvio baixo -> proximo da media. pouca variabilidade

desvio alto -> afastado da media. muita variabilidade

```python
np.std(arr6)
```

// dados estão dispersos

```python
arr7 = np.array([12, 15, 20, 19])
np.std(arr7)
```

// dados estão concentrados

**Variância**

mostra a dispersão dos valores em relação à media. é calculada como a média dos quadrados da diferença entre cada valor e a média

*fórmula*:

var = 1/n * ∑(xi - m)^2


```python
np.var(arr6)
```
